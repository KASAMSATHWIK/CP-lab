```python
#Backtracking Assignment 3: Subset Sum Problem Elaborated Scenario

#A person is planning expenses and wants to know if some combination of expenses exactly matches a fixed budget.




def subset_sum(arr, n, target):


    if target == 0:
        return True


    if n == 0 or target < 0:
        return False


    include = subset_sum(arr, n - 1, target - arr[n - 1])


    exclude = subset_sum(arr, n - 1, target)

    return include or exclude

n = int(input())
arr = list(map(int, input().split()))
target = int(input())


result = subset_sum(arr, n, target)


print(result)

```
