```python
# Bellman-Ford Algorithm

# Step 1: Read N and M
first_line = input().strip()
while first_line == "":
    first_line = input().strip()

N, M = map(int, first_line.split())

# Step 2: Read edges
edges = []
for _ in range(M):
    line = input().strip()
    while line == "":
        line = input().strip()
    u, v, w = map(int, line.split())
    edges.append((u, v, w))

# Step 3: Read Source
source_line = input().strip()
while source_line == "":
    source_line = input().strip()

S = int(source_line)

# Step 4: Bellman-Ford
INF = float('inf')
distance = [INF] * N
distance[S] = 0

# Relax edges N-1 times
for _ in range(N - 1):
    for u, v, w in edges:
        if distance[u] != INF and distance[u] + w < distance[v]:
            distance[v] = distance[u] + w

# Step 5: Check negative cycle
for u, v, w in edges:
    if distance[u] != INF and distance[u] + w < distance[v]:
        print("NEGATIVE CYCLE")
        break
else:
    for d in distance:
        print(d if d != INF else "INF")
```
