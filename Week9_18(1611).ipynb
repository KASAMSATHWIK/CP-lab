```python
def build_z_array(s: str) -> list:
    """
    Construct the Z-array for a given string.

    Args:
        s (str): Combined string (pattern + '$' + text)

    Returns:
        list: Z-array
    """
    n = len(s)
    z = [0] * n
    left, right = 0, 0

    for i in range(1, n):
        if i <= right:
            z[i] = min(right - i + 1, z[i - left])

        while i + z[i] < n and s[z[i]] == s[i + z[i]]:
            z[i] += 1

        if i + z[i] - 1 > right:
            left, right = i, i + z[i] - 1

    return z


def find_pattern(text: str, pattern: str) -> list:
    """
    Find all occurrences of pattern in text using Z-algorithm.

    Args:
        text (str): Log text
        pattern (str): Pattern to search

    Returns:
        list: List of starting indices where pattern is found
    """
    combined = pattern + "$" + text
    z = build_z_array(combined)

    result = []
    pattern_length = len(pattern)

    for i in range(len(z)):
        if z[i] == pattern_length:
            result.append(i - pattern_length - 1)

    return result


# ------------------ MAIN ------------------
if __name__ == "__main__":
    text = input("Enter log text: ")
    pattern = input("Enter pattern: ")

    matches = find_pattern(text, pattern)

    if matches:
        for idx in matches:
            print(f"Pattern found at index {idx}")
    else:
        print("Pattern not found")
```
